```python
import matplotlib.pyplot as plt
import pandas as pd

# 주어진 데이터
data = {'prediction': [0 ,1 ,0 ,1 ,1 ,1 ,1 ,0 ,2 ,1 ,0 ,2 ,2 ,2 ,2 ,2 ,0 ,2 ,2 ,1 ,2 ,2 ,1 ,0 ,2 ,2 ,2 ,2 ,2 ,2 ,1 ,2 ,1 ,2 ,1, 0 ,1 ,2 ,1 ,2 ,1 ,2 ,2 ,0 ,0 ,2, 1, 2, 1, 0],
        # 'label':      [2, 2, 2, 2, 2, 2, 2, 2, 1, 2, 2, 1, 1, 1, 1, 1, 2, 1, 1, 2, 0, 1, 2, 1, 2, 1, 1, 1, 1, 1, 1, 2, 1, 2, 1, 2, 1, 2, 1, 0, 2, 1, 2, 2, 0, 0, 2, 1, 0, 2, 0, 2]}
        'label':      [2 ,2 ,2 ,2 ,2 ,2 ,2 ,2 ,1 ,2 ,2 ,1 ,1 ,1 ,1 ,1 ,2 ,1 ,1 ,2 ,0 ,1 ,2 ,2 ,1 ,1 ,1 ,1 ,1 ,1 ,2 ,1 ,2 ,1 ,2 ,2 ,2 ,1 ,0 ,1 ,2 ,1 ,1 ,2 ,2 ,1 ,2 ,0 ,2 ,2]}

df = pd.DataFrame(data)

label_mapping = {0: 'ham', 1: 'smishing', 2: 'spam'}
df['prediction_label'] = 'pred: ' + df['prediction'].map(label_mapping).astype(str) + '\nlabel: ' + df['label'].map(label_mapping).astype(str)

# 예측값과 라벨 간의 관계 계산
# df['relation'] = df.apply(lambda row: f'{row["prediction"]}-{row["label"]}', axis=1)

# 관계의 개수 계산
relation_counts = df['prediction_label'].value_counts()

# 그래프 그리기
plt.figure(figsize=(10, 6))
relation_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Prediction-Label')
plt.ylabel('Count')
plt.title('Prediction-Label')
plt.xticks(rotation=45, ha='right')

plt.xticks(range(len(relation_counts.index)), relation_counts.index, rotation=0, ha='center')
plt.show()
```
